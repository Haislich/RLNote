```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch


# Function to create and display a step in the interaction
def plot_step(step):
    fig, ax = plt.subplots(figsize=(8, 4))

    # Set limits and remove axes
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 5)
    ax.axis("off")

    # Add rectangles for 'Environment' and 'Learning algorithm'
    rect_env = FancyBboxPatch(
        (0.5, 2.5),
        2,
        1,
        boxstyle="round,pad=0.1",
        facecolor="lightblue",
        edgecolor="black",
    )
    rect_algo = FancyBboxPatch(
        (7.5, 2.5),
        2,
        1,
        boxstyle="round,pad=0.1",
        facecolor="lightblue",
        edgecolor="black",
    )

    ax.add_patch(rect_env)
    ax.add_patch(rect_algo)

    ax.text(1.5, 3, "Environment", fontsize=12, ha="center", va="center")
    ax.text(8.5, 3, "Learning\nalgorithm", fontsize=12, ha="center", va="center")

    if step == 1:
        # Add first interaction arrow (x_t ~ Î¼)
        arrow1 = FancyArrowPatch(
            (2.5, 3), (7.5, 3), arrowstyle="->", mutation_scale=15, color="black"
        )
        ax.add_patch(arrow1)
        ax.text(5, 3.2, r"$x_t \sim \mu$", fontsize=12, ha="center")

    elif step == 2:
        # Add second interaction arrow (a_t)
        arrow2 = FancyArrowPatch(
            (7.5, 3), (2.5, 3), arrowstyle="->", mutation_scale=15, color="black"
        )
        ax.add_patch(arrow2)
        ax.text(5, 2.5, r"$a_t$", fontsize=12, ha="center")

    elif step == 3:
        # Add third interaction arrow (r_t := r(x_t, a_t))
        arrow3 = FancyArrowPatch(
            (2.5, 3), (7.5, 3), arrowstyle="->", mutation_scale=15, color="black"
        )
        ax.add_patch(arrow3)
        ax.text(5, 1.55, r"$r_t := r(x_t, a_t)$", fontsize=12, ha="center")

    # Show the plot
    plt.show()


# Generate the sequence of interaction steps
for step in range(1, 4):
    plot_step(step)
```
