```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plot (empirical means for each arm)
arms = [1, 2, 3]
means = [1.7, 3, 2.7]  # Empirical means for arms 1, 2, 3
errors = [0.3, 0.3, 0.3]  # Symmetric error bars representing confidence intervals
true_means = [1.5, 2.8, 2.9]  # True means \mu_i for each arm

# Create the plot with a larger figure size
fig, ax = plt.subplots(figsize=(8, 4))

# Plot error bars for each arm (confidence intervals)
ax.errorbar(
    arms,
    means,
    yerr=errors,
    fmt="o",
    color="black",
    ecolor="black",
    elinewidth=2,
    capsize=5,
    capthick=2,
    label=r"Empirical mean $\hat{\mu}_i$",
)

# Add labels to each arm showing the confidence interval formulas
for i, arm in enumerate(arms):
    mu_label = r"$\hat{\mu}_{" + str(i + 1) + r"}$"
    upper_ci_label = (
        r"$\hat{\mu}_{" + str(i + 1) + r"} + \sqrt{\frac{\ln(K/\delta)}{N}}$"
    )
    lower_ci_label = (
        r"$\hat{\mu}_{" + str(i + 1) + r"} - \sqrt{\frac{\ln(K/\delta)}{N}}$"
    )

    # Upper label (bounding box to avoid overlap)
    ax.text(
        arm,
        means[i] + errors[i] + 0.1,
        upper_ci_label,
        ha="center",
        va="bottom",
        fontsize=10,
        bbox=dict(facecolor="white", alpha=0.6),
    )

    # Lower label (bounding box to avoid overlap)
    ax.text(
        arm,
        means[i] - errors[i] - 0.1,
        lower_ci_label,
        ha="center",
        va="top",
        fontsize=10,
        bbox=dict(facecolor="white", alpha=0.6),
    )

# Set x-ticks to show the arms
ax.set_xticks(range(5))
ax.set_xticklabels([])
ax.set_yticks(range(5))
# ax.set_xticklabels([r"$\hat{\mu}_1$", r"$\hat{\mu}_2$", r"$\hat{\mu}_3$"])

# Add axis labels and title
ax.set_xlabel("Arms")
ax.set_ylabel("Empirical Means and Confidence Intervals")
ax.set_title("Confidence Intervals")

ax.legend(loc="lower right")

# Adjust layout to ensure labels fit well
plt.tight_layout()


# Show the plot
plt.show()
```
